```python
import requests
from bs4 import BeautifulSoup

# Step 1: Make an HTTP request to the website
url = 'https://www.vogue.com/fashion/trends'  # Replace with the actual URL
response = requests.get(url)

# Step 2: Check if the request was successful
if response.status_code == 200:
    # Step 3: Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Step 4: Find all div elements with a specific class
    target_class = 'BaseWrap-sc-gjQpdd RowWrapper-UmqTg iUEiRd HEhan CarouselRow-fiKuyQ yOvWk product__carousel'  # Replace with the actual class name
    divs = soup.find_all('div', class_=target_class)

    # Step 5: Extract and print the content of each div
    for div in divs:
        # Extract the title
        title = div.find('h2').get_text() if div.find('h2') else 'No title found'
        # Extract the link
        # link = div.find('a')['href'] if div.find('a') and 'href' in div.find('a').attrs else 'No link found'
        links = [a['href'] for a in div.find_all('a', href=True)]
        links_text = ", ".join(links) if links else 'No links found'

        # Extract the summary or description
        summary = div.find('p').get_text() if div.find('p') else 'No summary found'

        print(f"Title: {title}")
        print(f"Links: {links_text}")
        print(f"Summary: {summary}\n")
else:
    print(f"Error fetching the page: {response.status_code}")
```

```python
# scraping content for all secondary links

# url1 = final_secondary_link[1]

secondary_links_text = {}
for link in final_secondary_link:

  # Step 1: Make an HTTP request to the website
  # url = url1  # Replace with the actual URL
  response = requests.get(link)

  # Step 2: Check if the request was successful
  if response.status_code == 200:
      # Step 3: Parse the HTML content using BeautifulSoup
      soup = BeautifulSoup(response.content, 'html.parser')
      text = ''

      # Step 4: Find the outer div with the specified class
      outer_div_class = 'BodyWrapper-kufPGa jvNNFa body body__container article__body'  # Replace with the actual class name
      outer_div = soup.find('div', class_=outer_div_class)

      if outer_div:
          # Step 5: Find the inner div within the outer div
          inner_div_class = 'body__inner-container'
          inner_div = outer_div.find('div', class_=inner_div_class)

          if inner_div:
              # Step 6: Find all p tags within the inner div
              p_tags = inner_div.find_all('p')

              # text = ''
              # Step 7: Extract and print the content of each p tag
              for p in p_tags:
                text = text + '\n' + p.get_text()
                print(p.get_text())

          else:
              print("Inner div not found")
      else:
          print("Outer div not found")
      secondary_links_text[link] = text
  else:
      print(f"Error fetching the page: {response.status_code}")
```

## Storing scraped data into pdf

```python

class PDF(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 12)
        self.cell(0, 10, 'Myntra Fashion Chatbot', 0, 1, 'C')

    def footer(self):
        self.set_y(-15)
        self.set_font('Arial', 'I', 8)
        self.cell(0, 10, f'Page {self.page_no()}', 0, 0, 'C')

# Create PDF instance
pdf = PDF()
pdf.set_auto_page_break(auto=True, margin=15)
pdf.add_page()

# Set title font
pdf.set_font('Arial', 'B', 16)

# Loop through the dictionary and add content
for news in trends_desc_content.keys():
  for title, content in trends_desc_content[news].items():
    pdf.set_font('Arial', 'B', 16)
    pdf.cell(0, 10, title.encode('latin-1', 'replace').decode('latin-1'), 0, 1)
    pdf.ln(5)
    pdf.set_font('Arial', '', 12)
    pdf.multi_cell(0, 10, content.encode('latin-1', 'replace').decode('latin-1'))
    pdf.ln(10)


# Save the PDF
pdf.output('fashion_trending_news.pdf')
```
